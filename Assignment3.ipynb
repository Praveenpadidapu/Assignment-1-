#plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.  Label the x-axis as "X Values" and the y-axis as "Squared Values." o Add a title to the graph.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11)
y = x**2

plt.plot(x, y)
plt.xlabel('X values')
plt.ylabel('Squared Values')
plt.title('Square of X')
plt.show()
```

#Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi', 'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88]. o Add labels to the x-axis and y-axis and give the chart a title.

# Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
#with values [20, 30, 25, 15, 10].
# Display the percentage values on the chart and use a color palette for
#differentiation.

```python
import numpy as np
import matplotlib.pyplot as plt
data=[20,30,25,15,10]
index=['A','B','C','D','E']
plt.pie(data,labels=index,autopct='%1.1d%%')#gives integers
plt.show()
```

#Create a 5x5 matrix with random integers between 1 and 10 using
#np.random.randint() and visualize it as a heatmap using Sea

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
matrix=np.random.randint(1,11, size=(5,5))
sns.heatmap(matrix,annot=True,cmap='PuBuGn')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Heat map")
plt.show()
```

#Generate a pair of plots (bar charts or line plots) using the same data in
#both Matplotlib and Seaborn.
# Highlight the differences in the code, syntax, and aesthetics of the two
#libraries.

**Difference**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {'Students': ['Sai','Vamsi', 'Divij', 'Sritan', 'Sita'],
        'Marks': [85, 90, 78, 92, 88]}
df = pd.DataFrame(data)
df = df.set_index('Students')
sns.heatmap(df, annot=True, cmap='PuBuGn')
plt.xlabel('Students Name')
plt.ylabel('Marks Secured')
plt.title('Seaborn Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
Students=['Sai','Vamsi', 'Divij', 'Sritan', 'Sita']
Marks=[85, 90, 78, 92, 88]
plt.bar(Students,Marks)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart using Matplotlib")
plt.show()
```

#Manually create a small dataset with columns such as 'Age', 'Height',
